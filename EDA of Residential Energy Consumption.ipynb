# EDA 

```python
import pandas as pd
%matplotlib inline
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
import seaborn as sns
```

## Read in df

```python
df = pd.read_csv('Residential_Consumption_data.csv', index_col='Years', parse_dates=True)
```

## Plot Data

```python
df['Consumption'].plot(figsize=(12,8), legend=True, ylabel='Natural Gas Consumption (MCF)')
```

**Seeing this graph, there are clearly two trends. One from 1930-1972 and another one from 1972-2022. This begs the question, why? Was there a US policy change? Did we not have enough infrastructure in place? Was it not profitable enough to produce more natural gas? Energy demands have vastly increased from the 70s to today in the US. Did renewables and oil cover the energy demands?**

```python
df['Consumption'].loc["1930-12-31":"1972-12-31"].plot(figsize=(12,8), legend=True, ylabel='Natural Gas Consumption (MCF)', title='1930 - 1972 Natural Gas Consumption')
```

```python
df['Consumption'].loc["1972-12-31":"2022-12-31"].plot(figsize=(12,8), legend=True, ylabel='Natural Gas Consumption (MCF)', title='1972 - 2022 Natural Gas Consumption')
```

**These graphs clearly show two different trends. One looks like an exponential graph while the other looks stationary. I will do an analysis on three different data sets: the whole dataset, from 1930-1972, and from 1973-2022**

## Seasonal Decompose

```python
df_first_half = df.loc["1930-12-31":"1972-12-31"]
df_second_half = df.loc["1972-12-31":"2022-12-31"]
```

```python
seasonal_decompose(df).plot();
```

```python
seasonal_decompose(df_first_half).plot();
```

```python
seasonal_decompose(df_second_half).plot();
```

```python
sns.boxplot(df['Consumption'])
```

```python
sns.boxplot(df_first_half['Consumption'])
```

```python
sns.boxplot(df_second_half['Consumption'])
```
